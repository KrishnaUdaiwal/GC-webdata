```python
import re

#for x in sel_data['keywords_str']:
def split_keywords(list_of_keywords):
    keywords = []
    try:
        if ',' in list_of_keywords or ';' in list_of_keywords or '\n' in list_of_keywords:
            keywords = re.split(r',|;|\n', list_of_keywords)
            # Assuming list:
            assert type(keywords) is list
            for i in keywords:
                if len(i) <= 2:
                    keywords.remove(i)
                # if len(i) <= 2:
                #     keywords.remove(i)
                    # elif len(i) > 50:
                    #     print("POTENTIAL: ", i)
        elif type(list_of_keywords) is float:
            return None
        else:
            keywords = [list_of_keywords]
        #[print(len(i), "/", len(list_of_keywords)) for i in keywords]

        keywords = [i.strip().lower() for i in keywords]
        #print("\n", list_of_keywords, "######################", keywords)
        return keywords
    except:
        pass
sel_data['keywords_split'] = [split_keywords(x) for x in sel_data['keywords_str']]
```

```python
# Written using copilot
from neo4j import GraphDatabase

import config
remote = config.remote
localhost = config.localhost

# Define the connection details
def use_profile(profile):
    global uri, user, password
    uri = profile['uri']
    user = profile['user']
    password = profile['password']

# Define the Cypher queries to create nodes and relationships
create_host_query = "MERGE (:Host {name: $name})"
create_title_query = "MERGE (:Title {name: $name})"
create_keyword_query = "MERGE (:Keyword {name: $name})"
create_host_title_relationship_query = """
MATCH (h:Host {name: $host})
MATCH (t:Title {name: $title})
MERGE (h)-[:HAS_TITLE]->(t)
"""

create_title_keywords_relationship_query = """
MATCH (t:Title {name: $title})
MATCH (k:Keyword {name: $keyword})
MERGE (t)-[:HAS_KEYWORD]->(k)
"""

# Define a function to execute the queries for each row in sel_data
def create_nodes_and_relationships(tx):
    for _, row in sel_data.loc[:,['host','title','keywords_split']].iterrows():
        # Create a node for the host
        tx.run(create_host_query, name=row['host'])
        
        # Create a node for the title
        if not pd.isna(row['title']):
            tx.run(create_title_query, name=row['title'])
        
        # Create nodes for each keyword
        if row['keywords_split'] is not None:
            for keyword in row['keywords_split']:
                tx.run(create_keyword_query, name=keyword)

        # Create relationships between the host and title
        tx.run(create_host_title_relationship_query, host=row['host'], title=row['title'])

        # Create relationships between the title and keywords
        if row['keywords_split'] is not None:
            for keyword in row['keywords_split']:
                tx.run(create_title_keywords_relationship_query, title=row['title'], keyword=keyword)

```

```python
# Execute the queries within a transaction
use_profile(localhost)
# Connect to the Neo4j instance
driver = GraphDatabase.driver(uri, auth=(user, password))

with driver.session() as session:
    session.write_transaction(create_nodes_and_relationships)
```

```python
# Created with copilot

import concurrent.futures
import networkx as nx

# Create a new graph
G = nx.Graph()

# Add nodes for each keyword in keywords_all that is in sel_data['content']
def add_nodes(keyword):
    if keyword in sel_data['content'].str.lower().tolist():
        G.add_node(keyword)

# Add edges between nodes that appear in the same content
def add_edges(i, row1):
    for j, row2 in sel_data.iterrows():
        if i != j:
            for keyword1 in row1['keywords_split']:
                if keyword1 is not None and keyword1 in G.nodes():
                    for keyword2 in row2['keywords_split']:
                        if keyword2 is not None and keyword2 in G.nodes():
                            if keyword1 != keyword2 and keyword1 in row2['content'].lower() and keyword2 in row1['content'].lower():
                                G.add_edge(keyword1, keyword2)

with concurrent.futures.ThreadPoolExecutor() as executor:
    # Add nodes for each keyword in keywords_all that is in sel_data['content']
    futures = [executor.submit(add_nodes, keyword) for keyword in keywords_all]
    for future in concurrent.futures.as_completed(futures):
        pass

    # Add edges between nodes that appear in the same content
    futures = [executor.submit(add_edges, i, row1) for i, row1 in sel_data.iterrows()]
    for future in concurrent.futures.as_completed(futures):
        pass

# Draw the graph
nx.draw(G, with_labels=True)
```
